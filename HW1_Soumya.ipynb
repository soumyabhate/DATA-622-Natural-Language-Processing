Q1. Retrieve the following Web Page Use the requests library to fetch the content of https://www.visitmaryland.org/ What is the HTTP status code?

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin
```

```python
url="https://www.visitmaryland.org/"
header={"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64)"}

res=requests.get(url,headers=header)
print("Status Code is :",res.status_code)
```

Q2. Extract the Main Page Text Using BeautifulSoup, extract and print all the visible text from the page (ignore script and style tags).

```python
s=BeautifulSoup(res.text, "html.parser")  
for htmltags in s(["script", "style"]):
    htmltags.decompose()

txt = s.get_text(strip=True)  
print(txt)
```

Q3. Extract Headings from a Wikipedia Page Fetch the Wikipedia page for Natural Language Processing. Extract and print all the headings (h1, h2, h3) from the page.

```python
wiki_url="https://en.wikipedia.org/wiki/Natural_language_processing"

r=requests.get(wiki_url, headers=header)

soup=BeautifulSoup(r.text) 

print("\n H1 Text:\n")
for i in soup.select("h1"):
    print(i.get_text(strip=True))

print("\n H2 Text:\n")
for i in soup.select("h2"):
    print(i.get_text(strip=True))

print("\n H3 Text:\n")
for i in soup.select("h3"):
    print(i.get_text(strip=True))
```

Q4. Extract Links from a Page From the Wikipedia page above, extract and print all the URLs (href values) of links found on the page.

```python
links=set()
for a in soup.find_all("a", href=True):
    href=a.get("href")
    full=urljoin(wiki_url, href)
    links.add(full)

for u in sorted(links):
    print(u)
```

Q5. Extract and Save Text Extract the first paragraph (p) from the Wikipedia page and save it to a local text file called nlp_intro.txt.

```python
#wikipedia stores body in .mw-body-content
pText=None
for p in soup.select(".mw-body-content p"):
    text=p.get_text(' ',strip=True)
    if text:
        pText=text
        break
        
with open("nlp_intro.txt", "w", encoding="utf-8") as f:
    f.write(pText+'\n')
```
